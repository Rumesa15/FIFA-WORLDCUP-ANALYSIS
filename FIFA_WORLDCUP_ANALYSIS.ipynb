# DATA MANIPULATION:

Data manipulation involves cleaning , transforming and preparing the data for meaningful insights.

* Between Matches and players datasets: *
RoundID and MatchID  are common columns.
* Betwwen Matches and Players and Tournament datasets:*
Year is the Common Column


**1.Merge Matches and Players Datasets : **

```python
matches_players =pd.merge(matches , players , on=['RoundID','MatchID'], how='inner')
```

```python
matches_players
```

We have got the dataset with match details ( e.g, goals , stadium , attendance ) combined with player details (e.g., position , events)

** 2.Mergeing the Result with Tournament Dataset: **

```python
fifa_data = pd.merge(matches_players,worldcup, on='Year', how = 'inner')
```

```python
fifa_data
```

Now we have combined dataset with details from all three datasets.

**2.CALCULATE AGGREGRATES:**

Performing aggregations to exctract meaningful insights.


1. Calculate average number of players per match for each world cup year.
2. Analyzing attendance trends accross different Years.
1. Indentify the teams with the highest number of goals in each world cup.






* 1.Calculating average number of players per match for each world cup year. *

```python
avg_players_per_match = fifa_data.groupby('Year')['MatchID'].nunique().reset_index()
```

* 2.Analyzing attendance trends accross different Years.*

```python
print(fifa_data['Attendance_y'].dtype)
```

The data type of Attendance column is object and to perform statistical analysis like mean or median it returns errors so dataset column must have in float type. so cleaning the column of the dataset and converting the data type from object to folat.

1.Remove commas or non-numeric characters .Using str.replace() to clean the strings and converting them to numeric.

```python
fifa_data['Attendance_y'] = fifa_data['Attendance_y'].str.replace(',', '').str.strip()
```

2.Convert to numeric : converting the cleand column to numeric

```python
fifa_data['Attendance_y'] = pd.to_numeric(fifa_data['Attendance_y'] , errors='coerce')
```

3.Handling missing or invalid values :

```python
fifa_data['Attendance_y'].fillna(0,inplace=True)
```

```python
print(fifa_data['Attendance_y'].dtype)
```

Now the column is ready to be used for analysis.

```python
attendance_trends = fifa_data.groupby('Year')['Attendance_y'].mean().reset_index()
```

```python
attendance_trends
```

**3.Identify the teams with the highest number of goals in each worldcup:**

```python
top_team_goals= fifa_data.groupby(['Year','Home Team Name'])['Home Team Goals'].sum().reset_index()
```

```python
top_team_goals = top_team_goals.sort_values(by=['Year','Home Team Goals'],ascending=False)
```

```python
top_team_goals
```

**Filter data for specific scenarios:**

As we have merged the dataset now we can also filter the data for analyzing specific scenarios to make more detail interpretation  like

1. Matches with highest attendance
2. Top players in finals



```python
top_attendance_matches = fifa_data.sort_values(by='Attendance_y',ascending=False).head(10)
```

```python
top_attendance_matches
```

# **DATA VISUALISATION:**

Now that we have performed some analysis on the dataset and data mainuplation part has been done . We've got some idea about the patterns like attendance by players and some relationships of columns of one dataset to other.
Here is the visulaisation of the data so we get some concusions based on that

Using matplotlib and seaborn libraries provided by python

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**1.Attendance treds accross years:**
Analyze how the average attendance varied across different worldcups.Using line chart to show  attendance trends over time.

```python
plt.figure(figsize =(10,6))
sns.lineplot(data=attendance_trends,x='Year',y='Attendance_y' , color = 'blue')
plt.title('Average Attendance Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Average Attendance')
plt.show()
```

From the above line chart it interprets that the average attendance trends for FIFA World cup events from 1950-2000.There's a noticeable increase in attendance from 1950-1990,followed by sharp decline.

**2.Goals scored by teams:**

Identifying which teams scored the most goals in a specific worldcup year, by visualizing a horizontal bar chart for top teams based on goals.
let's for example taking year 2014 for account

```python
year_2014=top_team_goals[top_team_goals['Year']==2014]
```

```python
year_2014
```

```python
plt.figure(figsize = (12,6))
sns.barplot(data=year_2014, x='Home Team Goals', y ='Home Team Name', palette ='viridis', orient = 'h')
plt.title('Top Teams with Most Goals in 2014')
plt.xlabel('Number of Goals')
plt.ylabel('Team Name')
plt.show()
```

The above bar chart highlights germany leading with the most goals , followed by Brazil and Argentina. This provides insights into team performance during the 2014 FIFA WorldCup.

**3.Attendance by RoundID and Year:**
Identify patterns in attendance across different roundID and Years.
First step is to create a pivot table for attendance.

```python
attendance_heatmap = fifa_data.pivot_table(index='RoundID', columns ='Year' , values='Attendance_y', aggfunc='mean')
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(attendance_heatmap , annot=True , fmt='.0f', cmap='coolwarm')
plt.title('Attendance across  RoundID and Years')
plt.xlabel('years')
plt.ylabel('RoundID')
plt.show()
```

*The above heatmap displays FIFA WorldCup attendance across various years and rounds.Red areas indicate higher attendance,while blue areas show lower attendance,highlighting key trends in event popularity.*

**4.Distribution of goals:**
Understanding distribution of goals scored in matches.

```python
plt.figure(figsize=(10,6))
sns.histplot(fifa_data['Home Team Goals'], bins=15, kde=True, color='green')
plt.title('Distribution of Home Team Goals Scored per Match')
plt.xlabel('Goals Scored')
plt.ylabel('Number of matches')
plt.show()
```

*The above histogram shows that most FIFA WorldCup matches have 2-3 goals scored per match,with fewer  matches having higher or lower goal counts.This highlights common scoring patterns in WorldCup matches.*

**5.Matches with the highest attendance:**
Investigating matches with the highest attendance by  top attendance matches dataframe and plotting

```python
top_attendance_matches
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=top_attendance_matches, x='Attendance_y', y='Home Team Name', palette='viridis')
plt.title('Matches with the Highest Attendance')
plt.xlabel('Attendance')
plt.ylabel('Home Team Name')
plt.show()
```

The above bar chart highlights that matches involving England and Yugoslavia had the highest  attendance figures in FIFA WorldCup analysis, both reaching approximately 800 attendees. This underscores the significant fan engagement for these teams.
